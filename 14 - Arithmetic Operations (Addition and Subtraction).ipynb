```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image1 = cv2.imread("/Users/alireza/Coding/OpenCV_Learning/OpenCV_Learnings/Images/img.jpg",cv2.COLOR_BAYER_BG2GRAY)
image2 = cv2.imread("/Users/alireza/Coding/OpenCV_Learning/OpenCV_Learnings/Images/Stanford_University_Logo_full.png",cv2.COLOR_BAYER_BG2GRAY)
image2 = cv2.resize(image2, (image1.shape[1], image1.shape[0]))
#to have the in same size

#add_image = cv2.addWeighted(image1,0.2,image2,0.5,0.4)
weightedSum = cv2.addWeighted(image1, 0.5, image2, 0.4, 0)
plt.imshow(weightedSum)
cv2.waitKey(1000)
# 🚨 image1 and image2 do not have the same size or number of channels, which is required for cv2.addWeighted().

```
