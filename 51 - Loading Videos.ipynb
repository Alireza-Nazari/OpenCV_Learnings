## 51 - Loading Videos

📹 Load the Video File
Use cv2.VideoCapture(). You can pass a filename or camera index (like 0 for webcam).

cv2.VideoCapture(0): Means first camera or webcam.

cv2.VideoCapture(1):  Means second camera or webcam.

cv2.VideoCapture("file name.mp4"): Means video file

```python
import cv2
import matplotlib.pyplot as plt
cap = cv2.VideoCapture('/Users/alireza/Coding/OpenCV_Learning/OpenCV_Learnings/Videos/drummer.mp4')  # Replace with your file path
# Or: cap = cv2.VideoCapture(0)      # For webcam
```

🔁 3. Read and Display Frames in a Loop

✅ ret (Boolean)

True if the frame was successfully read.

False if the video ended or an error occurred (e.g., file not found, corrupt video).

🖼️ frame (NumPy array)

This is the actual image frame (a single still from the video).



```python
while cap.isOpened():
    ret, frame = cap.read()  # ret: success flag, frame: image

    if not ret:
        print("can't fetch the frame.")
        break

    cv2.imshow('video',frame)  # Show frame

    if cv2.waitKey(25) & 0xFF == ord('q'):  # Press 'q' to quit
        break
```

### ✅ Show 10 Video Frames with Matplotlib
Here’s how you can extract and display 10 frames from a video using OpenCV and Matplotlib:

```python
cap = cv2.VideoCapture('/Users/alireza/Coding/OpenCV_Learning/OpenCV_Learnings/Videos/drummer.mp4')

frames = list()
count = 0

# We want to read 12 frames and show them with matplotlib

while count < 12 :
    ret,frame = cap.read()
    if ret is None:
        break
    else:
        frame = cv2.cvtColor(frame , cv2.COLOR_BGR2RGB)
        frames.append(frame)
        count = count + 1

cap.release()

plt.figure(figsize=(15,6))
for i in range(len(frames)):
    plt.subplot(3,4,i+1)
    plt.imshow(frames[i])
    plt.axis('off')
plt.tight_layout()
plt.show()

```
