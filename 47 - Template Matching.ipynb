```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
template = cv2.imread('/Users/alireza/Coding/OpenCV_Learning/OpenCV_Learnings/Images/tmp.png')
image = cv2.cvtColor(template , cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
# Load input image and convert to grayscale
image = cv2.imread('/Users/alireza/Coding/OpenCV_Learning/OpenCV_Learnings/Images/template.png')
image = cv2.cvtColor(image , cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Load Template image
template = cv2.imread('/Users/alireza/Coding/OpenCV_Learning/OpenCV_Learnings/Images/tmp.png',0)

result = cv2.matchTemplate(gray, template, cv2.TM_CCOEFF)
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(result)

#Create Bounding Box
top_left = max_loc
bottom_right = (top_left[0] + 50, top_left[1] + 50)
cv2.rectangle(image, top_left, bottom_right, (0,0,255), 5)

plt.imshow( image)
```
