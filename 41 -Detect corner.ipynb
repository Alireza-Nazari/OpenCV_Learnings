## 41 -Detect corner

Syntax : 

cv2.goodFeaturesToTrack(image, maxCorners, qualityLevel, minDistance[, corners[, mask[, blockSize[, useHarrisDetector[, k]]]]])

```python
import numpy as np 
import matplotlib.pyplot as plt 
import cv2
```

```python
img = cv2.imread('/Users/alireza/Coding/OpenCV_Learning/OpenCV_Learnings/Images/corner1.png')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray,cmap='gray')
```

```python
# detect corners with the goodFeaturesToTrack function.
corners = cv2.goodFeaturesToTrack(gray, 28, 0.01, 10)
corners = np.int0(corners)

# we iterate through each corner, 
# making a circle at each point that we think is a corner.
for i in corners:
    x, y = i.ravel() #.ravel() function is used to flatten a multi-dimensional array into a 1D array.
    cv2.circle(img, (x, y), 4, 255, -1)

plt.imshow(img), plt.show()
```
