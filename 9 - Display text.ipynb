## 9️⃣ - Displaying text
It is also an in-place operation that can be done using the putText() method of OpenCV module. It takes in 7 arguments:

Image. /
Text to be displayed. /
Bottom-left corner co-ordinates, from where the text should start. /
Font. /
Font size. /
Color (BGR format). /
Line width. /

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("/Users/alireza/Coding/OpenCV_Learning/OpenCV_Learnings/Images/img.jpg")
texted_image = cv2.putText(image , 'text on image' , (500, 550), cv2.FONT_HERSHEY_SIMPLEX, 4 ,(255, 0, 0), 2)
plt.imshow(texted_image)
# if i dont puth this waitkey it will not show me the result
#  waitkey(0) to hold the image window on the screen by the specified number of seconds, o means 
# till the user closes it, it will hold GUI window on the screen.
# destroyAllWindows()
# Delete image window from the memory after displaying using destroyAllWindows() function.
```
