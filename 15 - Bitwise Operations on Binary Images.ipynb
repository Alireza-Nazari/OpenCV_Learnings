## 15 - Bitwise Operations on Binary Images

### NOTE: The Bitwise operations should be applied on input images of same dimensions.



```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread("/Users/alireza/Coding/OpenCV_Learning/OpenCV_Learnings/Images/black.png")
img2 = cv2.imread("/Users/alireza/Coding/OpenCV_Learning/OpenCV_Learnings/Images/white circle.png")
And = cv2.bitwise_and(img1,img2,mask = None)
plt.imshow(And)
cv2.waitKey(1000)
```
